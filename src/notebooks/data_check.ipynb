### Feature Engineering

```python
df["pickup_date"] = df["tpep_pickup_datetime"].dt.date
```

```python
df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour
```

```python
df["is_weekend"] = df["tpep_pickup_datetime"].dt.dayofweek.isin([5, 6])
```
